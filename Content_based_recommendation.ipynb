```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df=pd.read_csv('https://query.data.world/s/uikepcpffyo2nhig52xxeevdialfl7')
df.head()
```

```python
df = df[['Title','Genre','Director','Actors','Plot']]
df.head()
```

```python
df['Actors']=df['Actors'].map(lambda x:x.split(',')[:3])
```

```python
df['Actors']
```

```python
df['Genre'] = df['Genre'].map(lambda x: x.lower().split(','))
```

```python
df['Genre']
```

```python
df['Director']=df['Director'].map(lambda x: x.split())
```

```python
df['Director']
```

```python
for ind,rows in df.iterrows():
  rows['Actors']=[i.lower().replace(' ','') for i in rows['Actors']]
  rows['Director']=''.join(rows['Director']).lower()
```

```python
df.head()
```

```python
import rake_nltk
from rake_nltk import Rake
import nltk
nltk.download('stopwords')
nltk.download('punkt')
```

```python
df['key_words']=''
for index, row in df.iterrows():
    plot = row['Plot']
    r = Rake()
    r.extract_keywords_from_text(plot)
    key_words_dict_scores = r.get_word_degrees()
    row['key_words']=list(key_words_dict_scores.keys())
```

```python
df.drop('Plot',inplace=True,axis=1)
```

```python
df.head()
```

```python
df['Genre'] = df['Genre'].apply(lambda x: ' '.join(x))
df['Actors'] = df['Actors'].apply(lambda x: ' '.join(x))
df['key_words'] = df['key_words'].apply(lambda x: ' '.join(x))
```

```python
df['bag_of_words']=df['Genre'] + ' ' + df['Director'] + ' '+ df['Actors'] + ' ' + df['key_words']
```

```python
df.drop(['Genre','Director','Actors','key_words'],axis=1,inplace=True)
```

```python
df.head()
```

```python
count=CountVectorizer()
dtm=count.fit_transform(df['bag_of_words'])
dtm
```

```python
# generating the cosine similarity matrix
cosine_sim = cosine_similarity(dtm, dtm)
pd.DataFrame(cosine_sim)
```

```python
film_name=df['Title']
film_name
```

```python
def recommendation(title,cosine_sim=cosine_sim):
  recommended_movies=[]
  try:
    idx=film_name[film_name==title].index[0]
    score_series=pd.Series(cosine_sim[idx]).sort_values(ascending=False)
    top_10=list(score_series.iloc[1:11].index)
    for i in top_10:
      recommended_movies.append(film_name[i])
    return recommended_movies
  except:
    print('The movie not Found!!!')


```

```python
recommendation("The Shawshank Redemption")
```
